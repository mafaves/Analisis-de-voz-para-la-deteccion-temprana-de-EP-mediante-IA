# Plot audio data

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_waveform_middle_segment(audio_data, sr, label, patient_type, window_ms=20):
    """Plot a waveform for a 40 ms window from the middle of the audio."""
    # Calculate the total length of the audio in samples
    total_samples = len(audio_data)
    
    # Calculate the number of samples for the window (40 ms = 0.04 seconds, that's why we divide by 1000)
    window_samples = int(window_ms * sr / 1000)  # 40 ms in samples
    
    # Find the middle point of the audio
    middle_sample = total_samples // 2
    
    # Extract the window around the middle point
    start_sample = middle_sample - window_samples // 2
    end_sample = middle_sample + window_samples // 2
    middle_audio_segment = audio_data[start_sample:end_sample]
    
    # Create time vector for the extracted window
    time = np.linspace(0, window_ms / 1000, num=window_samples)
    
    # Plot the waveform
    plt.figure(figsize=(10, 4))
    plt.plot(time, middle_audio_segment, label=f'Patient Type: {patient_type}, Label: {label}')
    plt.xlabel('Time (seconds)')
    plt.ylabel('Amplitude')
    plt.title(f'Waveform (Middle {window_ms} ms) for Patient Type: {patient_type}, Label: {label}')
    plt.legend()
    plt.grid(True)
    plt.show()
```

### Training data

```python
# Initialize lists to store waveforms and labels for 1-second chunks
waveforms_chuncks = []
chunk_labels = []

# Loop through each 1-second chunk and collect waveforms and labels
for i, chunk in enumerate(preprocessed_chunks_data_np):
    try:
        # Convert each chunk (numpy array) to a Waveform object from Surfboard
        # Assuming the sample rate is 16kHz
        waveform = Waveform(signal=chunk, sample_rate=16000)  # Load chunk as a waveform
        
        # Append the waveform and label to their respective lists
        waveforms_chuncks.append(waveform)
        chunk_labels.append(labels_chunks_np[i])
    
    except Exception as e:
        print(f"Error processing chunk {i}: {e}")

# Now, extract features for all 1-second waveforms
features_training_df = extract_features(
    waveforms=waveforms_chuncks, components_list=components_list, statistics_list=statistics_list
)
```

```python
from scipy.stats import pearsonr
import pandas as pd

# Initialize dictionaries to store correlations and p-values for HC and PD groups
correlation_HC = {}
correlation_PD = {}

# Compute correlations for Healthy Control group (HC)
for feature in dataset1_HC_scaled_subset.columns:
    corr, p_value = pearsonr(dataset1_HC_scaled_subset[feature], dataset2_HC_scaled_subset[feature])
    correlation_HC[feature] = {'Correlation': corr, 'p-value': p_value}

# Compute correlations for Parkinson's Disease group (PD)
for feature in dataset1_PD_scaled_subset.columns:
    corr, p_value = pearsonr(dataset1_PD_scaled_subset[feature], dataset2_PD_scaled_subset[feature])
    correlation_PD[feature] = {'Correlation': corr, 'p-value': p_value}

# Convert results to pandas DataFrames for easier comparison and visualization
correlation_HC_df = pd.DataFrame(correlation_HC).T
correlation_PD_df = pd.DataFrame(correlation_PD).T

# Display the results
#print("Healthy Control Group Correlations")
#print(correlation_HC_df)

# print("\nParkinson's Disease Group Correlations")
# print(correlation_PD_df)
```
